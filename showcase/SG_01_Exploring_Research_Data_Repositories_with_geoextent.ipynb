## About Geoextent

Geoextent is a Python library for extracting geospatial extent of files and directories with multiple data formats. The library mostly relies on GDAL and text-scrapping techniques in order to extract the geospatial and temporal features of common file types storing geospatial information. The supported file formats include both vector (e.g. GeoPackage, Shapefile, GeoJSON) and raster (TIFF, GeoTIFF, JPEG) file types. The user can extract the bounding box (bbox) and/or time box (tbox) of the file.  

### Geoextent for individual files 

```python
# File of interest
local_filepath = "../tests/testdata/shapefile/ifgi_denkpause.shp"
# Geoextent extraction
geoextent_file = geoextent.fromFile(filepath = local_filepath,bbox=True,tbox=True)

# Print output

print(local_filepath,"\n","Bbox:",geoextent_file['bbox'],"\n",geoextent_file['crs'], "EPSG \n",
     "Tbox:",geoextent_file['tbox'])
```

### Geoextent for multiple files

```python
# Directory of interest
local_directory_path = "../tests/testdata/folders/folder_two_files"
# Geoextent extraction
geoextent_directory = geoextent.fromDirectory(path = local_directory_path,bbox=True,tbox=True,details=True)

# Print output

print(local_directory_path,"\n","Bbox:",geoextent_directory['bbox'],"\n",geoextent_directory['crs'], "EPSG \n",
     "Tbox:",geoextent_directory['tbox'])
```

## Case study

Zenodo is a data repository that stores different types of publication materials as unique `records`. Zenodo API allows the discovery of records by dozens of different parameters store in the metadata of the records including `title`,`doi`, `keywords` and `locations`. Even though geometadata is available as `locations` parameters, this options seems to be limited to only a small portion of records. In this sense, we would like to study the amount of records with geomatadata available in Zenodo and assess if by using **geoextent** we could provide the missing geometadata of some Zenodo records.


### Zenodo geometadata

The first step for this analysis is determining the current state of geometadata in the Zenodo records. As we are interested in analyzing some of these records with **geoextent**, only Open Access repositories are taken into account. For this purpose we are going to use the Zenodo __[spatial search](https://zenodo.org/api/records/?bounds=-180,-90,180,90)__ function available in their API.  This function requires the user to define an interest area (bounding box) to extract the records within this zone. Zenodo's geospatial information of a record is store in the `locations` property in the metadata. Each record could have multiple locations, but a location is limited to a single point. This point include a `lat`, `lon` and `place` features.

In the following cell we extract the proportion of open access Zenodo records with this `locations` property available. In this sense, is important to notice that Zenodo records include 9 different categories of publication types: `poster`, `presentation`, `dataset`, `image`, `video`, `software`, `lesson`, `physicalobject` and `other`. 

```python
## Parameters for search
world_bounds = "-180,-90,180,90"

# Type of Records
types_of_records = ["publication","poster", "presentation", "dataset", "image",
                    "video", "software", "lesson", "physicalobject","other"]

# Dictionary with search parameters
search_param = {}

# Zenodo search parameter for open records
search_param["open_record"] =  {"access_right":"open"}

# Zenodo search parameter for open records with geometadata
search_param["open_record_geometa"] =  dict(search_param["open_record"], bounds= world_bounds)

# Zenodo search parameter for open records with and without geometadata by type of record
for record_type in types_of_records:
    open_record = 'open_'+ record_type
    open_geometadata = open_record + "_geometa"
    search_param[open_record] = {"access_right":"open","type":record_type}
    search_param[open_geometadata] = {"access_right":"open","type":record_type,"bounds":world_bounds}
    
## Extract the number of records by search_parameter

num_of_records_by_search = {}

for type_of_zenodo_record in list(search_param.keys()):
    num_of_records_by_search[type_of_zenodo_record] = zenodo_api.get_number_of_records(search_param[type_of_zenodo_record])
    
```

Base on the information extracted from Zenodo's API, there are currently **{{'{:,}'.format(num_of_records_by_search['open_record'])}}** open records. From all of the open records **{{'{:,}'.format(num_of_records_by_search["open_record_geometa"])}}** have geospatial metadata. This means that **{{'{:.2f}'.format(num_of_records_by_search["open_record_geometa"]/num_of_records_by_search['open_record']*100)}}%** open Zenodo records are findable throught geospatial search. Among the records with geospatial metadata, **{{'{:.2f}'.format(num_of_records_by_search["open_publication_geometa"]/num_of_records_by_search["open_record_geometa"]*100)}}%** are`publications`({{'{:,}'.format(num_of_records_by_search['open_publication_geometa'])}} records) and **{{'{:.2f}'.format(num_of_records_by_search["open_dataset_geometa"]/num_of_records_by_search["open_record_geometa"]*100)}}%** are of `dataset` type ({{'{:,}'.format(num_of_records_by_search["open_dataset_geometa"])}} records). Even for these 2 categories with most of the records, 
only **{{'{:.2f}'.format(num_of_records_by_search["open_publication_geometa"]/num_of_records_by_search["open_publication"]*100)}}%** of open `publications` and **{{'{:.2f}'.format(num_of_records_by_search["open_dataset_geometa"]/num_of_records_by_search["open_dataset"]*100)}}%** open `dataset` include geospatial metadata.

To evaluate if we can use **geoextent** to increase the proportion of records with spatial metadata we use the result of a Zenodo API term search to create a list of records to analyze. In this particular search, we use the term *geo* and limit it to only `dataset` records smaller than 75 mb. From these records we extract basic metadata (e.g Title, DOI).

```python
# Parameters search for datasets with 'geo' term
geo_datasets_param = dict(search_param["open_dataset"],q="geo")
df_geo_zenodo_api = pd.DataFrame.from_dict(zenodo_api.get_info_records(geo_datasets_param),orient= 'index')
df_geo_zenodo_api = df_geo_zenodo_api[df_geo_zenodo_api["size_mb"]<75].copy()
```

## Analysis

After the extraction of geographical and temporal extent of zenodo repositories we have information for **repositories** and **files**. For that reason we are going to load the results of the geoextent extraction and slip it into two different tables. One table with the information of Records (repositories) and other table for individual files.

```python
df_geo = pd.read_csv("geoextent_extraction_geo.csv", index_col=0)

def wkt_loads(x):
    try:
        return wkt.loads(x)
    except:
        return None

df_geo['bbox'] = df_geo.bbox.apply(wkt_loads)
gdf_repository = gpd.GeoDataFrame(df_geo[df_geo['handler'] == "geoextent:0.7.0"],geometry ="bbox",crs = 4326)
gdf_repository.drop(['filename'], axis = 1,inplace = True)
gdf_files = gpd.GeoDataFrame(df_geo[df_geo['handler'] != "geoextent:0.7.0"],geometry ="bbox",crs = 4326)
```

### Geoextent extraction

There are three possible types of results after an extraction by Zenodo records: 1) A successful extraction information by geoextent, 2) no information extracted or 3) a failure ("error") of geoextent during the extraction.

```python
# Extract statistics for geospatial extraction 
num_records = len(gdf_repository)
per_records_geoextent = sum(gdf_repository.bbox.is_valid)/num_records*100
per_records_no_geoextent = (sum(gdf_repository.format == "repository")/num_records*100) - per_records_geoextent
per_records_failure = sum(gdf_repository.format == "repository_error")/num_records*100

# Extract statistics for temporal extraction
per_records_temp = (num_records-sum(gdf_repository.tbox.isnull()))/num_records*100
per_records_no_temp = (sum(gdf_repository.tbox.isnull())-sum(gdf_repository.format == "repository_error"))/num_records*100

# Store results
rec_successful = [per_records_geoextent,per_records_temp]
rec_no_extraction = [per_records_no_geoextent,per_records_no_temp]
rec_failure =  [per_records_failure,per_records_failure]

## Plot
fig, ax = plt.subplots(figsize=(15,3))

# Plot configuration
ext_type = ["Geospatial \n extraction","Temporal \n extraction"]
plt.title("Geoextent extraction of records by parameter")
plt.barh(ext_type, rec_successful,color='g',label ='Successful')
plt.barh(ext_type, rec_no_extraction,color='grey',label ='No extraction',left = rec_successful)
plt.barh(ext_type, rec_failure,color='red',label ='Geoextent failure',left = [sum(x) for x in zip(*[rec_successful,rec_no_extraction])])
plt.xlabel('(%) Percentage')
fig_num = 1
plt.annotate('Figure '+ str(fig_num)+'. Proportion of geoextent extraction status for zenodo records',
             (0,0), (200, -40), xycoords='axes fraction', weight='bold', textcoords='offset points', va='center')
plt.annotate('Extracted with geoextent \nhttps://o2r.info/geoextent/', (0,0), (710,-20), xycoords='axes fraction', textcoords='offset points', va='top')
fig_num +=1
plt.legend(loc=5)
plt.show()
```

From a total of **{{num_records}}** Zenodo records analyzed we were able to extract for **{{'{:.2f}'.format(per_records_geoextent)}}%** their geospatial component and for **{{'{:.2f}'.format(per_records_temp)}}%** their temporal component. Geoextent did not extract the geospatial and temporal extent in **{{'{:.2f}'.format(per_records_no_geoextent)}}%** and **{{'{:.2f}'.format(per_records_no_temp)}}%** of the cases respectively. This means that all the files of those records were analyzed, but no geospatial or temporal component was found. Finally the extraction with geoextent failed (crash) due to undetermined reasons in **{{'{:.2f}'.format(per_records_failure)}}%** of the records.

A similar analysis is possible for the individual files. In this case, only two outputs are possible: either the extraction is successful or there is no extraction at all. First, as geoextent stores the combined output of folders and zipfiles, we need to remove this type of registers to avoid overestimating the percentage of extraction.

```python
# Extract only files (no folders or zipfile)
uni_files_gdf = gdf_files[~gdf_files.format.isin(["folder","zip"])].copy().reset_index(drop=True)
uni_files_gdf.format = uni_files_gdf.format.str.lower()

# Extract statistics for geospatial extraction 
num_files = len(uni_files_gdf)

# Percentage files with valid bbox (i.e valid geometry) from total number of files
per_files_geoextent = sum(uni_files_gdf.bbox.is_valid)/num_files*100
# Percentage files without valid bbox (i.e no geometry) from total number of files
per_files_no_geoextent = 100-per_files_geoextent

# Extract statistics for temporal extraction
per_files_temp = (num_files-sum(uni_files_gdf.tbox.isnull()))/num_files*100
per_files_no_temp = 100-per_files_temp

# Record results
files_successful = [per_files_geoextent,per_files_temp]
files_no_extraction = [per_files_no_geoextent,per_files_no_temp]

# Plot
status = ["Successful","No extraction"]
fig, ax = plt.subplots(figsize=(15,3))
plt.title("Geoextent extraction of files by parameter")
plt.barh(ext_type, files_successful,color='g',label ='Successful')
plt.barh(ext_type, files_no_extraction,color='grey',label ='No extraction',left = files_successful)
plt.xlabel('(%) Percentage of files')
plt.xlim(0,100)
plt.legend(status,loc=5)
plt.annotate('Figure '+ str(fig_num)+'. Proportion of success and no extraction by individual files', (0,0), (250, -40), xycoords='axes fraction', weight='bold', textcoords='offset points', va='top')
plt.annotate('Extracted with geoextent \nhttps://o2r.info/geoextent/', (0,0), (710,-20), xycoords='axes fraction', textcoords='offset points', va='top')
fig_num+=1
plt.show()
```

From a total of **{{'{:,}'.format(num_files)}}** files analyzed, **{{'{:.2f}'.format(per_files_temp)}}%** had a successful temporal extraction and **{{'{:.2f}'.format(per_files_geoextent)}}%** a geospatial extraction. Files with no extraction of temporal or geospatial extraction are a combination a different types of files. For instance, the zenodo records could include information that does not necessarly includes (or require) temporal or geoespatial features as laboratory results, code or text-based documents. Similarly, even for file formats designed to store geospatial data, there is risk of ambiguity or lack of information regarding the coordinate reference systems. Lastly, even thought geoextent relies on GDAL that supports hundreds of raster and vector drivers, there could be file formats not supported by geoextent. 

In this sense, we took those files with known formats to potentially store geospatial information and classified them as files with potential geoextent. This group gives us a better idea of the proportion of geospatial information that could be extracted using geoextent from zenodo records. 

```python
# List of potential file formats with geospatial information
potential_geo_formats = ['csv','nc','tif', 'dbf',
                               'shp', 'shx', 'gpkg', 'geojson', 
                               'png','gml','xml','pdf','jpeg','jpg','qpj',
                               'png','json']


# Extract total number of files by zenodo record 
total_files = uni_files_gdf.zenodo_record_id.value_counts()

# Extract total number of potential file with geoextent by zenodo record
files_potential_geo_formats = uni_files_gdf['format'].isin(potential_geo_formats).groupby(uni_files_gdf['zenodo_record_id']).sum()

# Extract total number of files with potential geospatial information
total_files_potential = sum(files_potential_geo_formats)

# Extract percentage of files with successful geoextent extraction from the potential
per_total_geoextent_potential = sum(uni_files_gdf.bbox.is_valid)/total_files_potential*100

## Plot
fig, ax = plt.subplots(figsize=(15,3))
# Plot configuration
plt.title("Geoextent extraction status by potential files with geoextent")
plt.barh("Geoextent \n extraction",per_total_geoextent_potential,color='g',label ='Successful extraction')
plt.barh("Geoextent \n extraction",100-per_total_geoextent_potential,color='blue',label ='No extraction',left = per_total_geoextent_potential)
plt.xticks(np.arange(0,101, 5.0))
plt.annotate('Figure '+str(fig_num)+'. Distribution of files by Zenodo record', (0,0), (300, -40), xycoords='axes fraction', weight='bold', textcoords='offset points', va='top')
plt.annotate('Extracted with geoextent \nhttps://o2r.info/geoextent/', (0,0), (710,-20), xycoords='axes fraction', textcoords='offset points', va='top')
fig_num +=1
plt.legend()
plt.show()
```

From a total of **{{'{:,}'.format(num_files)}}** files analyzed there are **{{'{:,}'.format(total_files_potential)}}** files with known formats to potentially store geospatial information. From these files, only **{{'{:.2f}'.format(per_total_geoextent_potential)}}%** have a successful geoextent extraction (i.e BBOX extracted) and **{{'{:.2f}'.format(100-per_total_geoextent_potential)}}%** resulted in no extraction. The distribution of these files among the repositories is important to explore the percentages of zenodo records that could have a sucessful geoextent extraction (i.e at least one file with sucessful extraction).

In order to understand what are the problematic formats to extract geospatial information we compute its percentage of extraction success.

```python
uni_files_gdf['format'].value_counts()
files_valid = uni_files_gdf[uni_files_gdf.geometry.is_valid].copy().reset_index(drop=True)

d1 = pd.DataFrame({"Total":uni_files_gdf['format'].value_counts(),
                  "geoextent_success":files_valid['format'].value_counts()})
d1.dropna(inplace=True)
d1['perc_success'] = (d1['geoextent_success']/d1['Total'])*100

labels = d1.index.values
success = d1['perc_success']
failure = 100 - d1['perc_success']

color_success = "#009900"
color_failure = "#FF3333"
fig, ax = plt.subplots(figsize=(15,4))
width = 0.35
plt.bar(labels, success, width, label='Successful',color= color_success)
plt.bar(labels, failure, width, bottom=success, color = color_failure,label='No extraction')
plt.ylabel('Percentage')
plt.xlabel('File format')
plt.title('Geoextent extraction success by supported file type',size = 20)
plt.ylabel('Percentage')

count = 0
for bar in ax.patches: 
    displace =-1
    color = color_failure
    pos = 90
    if bar.get_y() == 0:
        n = d1['Total'][count]
        count+=1
        displace = 3
        color= color_success
        pos = 0
        plt.annotate("n= "+str(n),  
                   (bar.get_x() + (bar.get_width()/2),  
                   50), ha='center', va='center', 
                   size=10, xytext=(0, 0),color='black', 
                   textcoords='offset points')
    plt.annotate(format(bar.get_height()/100, ".1%"),  
                   (bar.get_x() + (bar.get_width()*1.5),  
                   pos), ha='center', va='center', 
                   size=10, xytext=(0, 8),color=color, 
                   textcoords='offset points')
    
plt.legend(loc='center right', bbox_to_anchor=(1.1, 0.5))
plt.annotate('Figure '+str(fig_num)+'. Distribution of files by Zenodo record', (0,0), (280, -40), xycoords='axes fraction', weight='bold', textcoords='offset points', va='top')
plt.annotate('Extracted with geoextent \nhttps://o2r.info/geoextent/', (0,0), (710,-20), xycoords='axes fraction', textcoords='offset points', va='top')
fig_num +=1
plt.show()
```

First, we can observe that fileformats which main purpose is to store geospatial information have a high success rate of extraction. For example, geojson files (json format for geospatial information) have a 100% sucess rate and .tif, .shp and .dbf vary between 80 and 97% of success rate. However,this is not always the case as for geopackage (gpkg), an open format for geospatial information, that have a lower success rate of 40%. This result could be atribute to the small sample (n=5) and errors in the formating of files as those present in other geospatial file formats(e.g **{{'{:.2f}'.format(100-d1.loc['shp'].perc_success)}}%** failure in shp files).

Second, the results suggest that ambiguous formats (i.e those that do not neccesarly have geospatial information) have a high percentage of no extractions. For csv files, geoextent could not extract geospatial information in **{{'{:.2f}'.format(100-d1.loc['csv'].perc_success)}}%** files. In order to determine what percentage correspond to information without geospatial information (e.g laboratory result) and those with geospatial information but unsuccessful extraction, we would need to evaluate each file one by one. However, what is relevant for us is that at least **{{'{:.2f}'.format(d1.loc['csv'].perc_success)}}%** of csv files have extractable geospatial information. A similar result is obtained for .png files with a **{{'{:.2f}'.format(d1.loc['png'].perc_success)}}%** extraction success rate. This means, that we can not discard ambiguous formats as potential sources of geometada.
