## Purpose

This notebook presents [geoextent](https://o2r.info/geoextent/index.html), a Python library for reliably extracting the geospatial and temporal extents of files, directories, and repository records. The geospatial and temporal metadata of research data could greatly benefit the discovery of relevant and related datasets, but it is underused in scientific data repositories with the exception of specialised repositories. However, much more scientific disciplines collect data and publish work that has some temporal or spatial relation. The library geoextent presented in this notebook helps to understand the potential of extract information from files shared in data repositories, and may be used to integrate geospatial and temporal metadata into repository infrastructures.

## Technical contributions

The geoextent library is a wrapper around the most commonly used software for geospatial data loading and saving, GDAL (GDAL/OGR contributors, 2021).
The main contribution is the ease of use of extracting discovery metadata from data files using GDAL, the handling of most common cases with defaults to support automation, the aggregation of extents for multiple files or directories, and the integration of retrieval functions for common scientific data repositories.
This notebook relies on geoextent version 0.7.TODO (Nüst, Garzón & Qamaz, 2021) and some helper functions shared next the notebook file.
This notebook is developed for Python 3.6+ and a standard [Jupyter](https://jupyter.org/) environment (Kluyver et al., 2016). Some cells require a stable connection to the [Zenodo API](https://developers.zenodo.org/).

## Methodology

We performed a case study of Zenodo records to explore the potential of automatically extractable geospatial metadata in research data repositories.
Furthermore, the case study validated the features of geoextent and improved the automated handling of data types.
First, a set of records based on the search term 'geo' and below a record size limit are downloaded and metadata are extracted with geoextent.
Results are stored in a local GeoPackage file and then analysed.
We determine the percentage of records where geospatial metadata can be automatically extracted and render the extracted geospatial and temporal metadata for visual inspetion.
Second, a few Zenodo records include the spatial extent. We compare these metadata, which are of unknown origin, with the metadata extracted by geoextent.

## geoextent usage

### Supported file types
geoextent supports a subset of the file formats supported by GDAL and uses text-scrapping techniques in order to extract the geospatial and temporal features of common file types storing geospatial information.
This subset includes the most commonly used formats and ensures a stable handling of edge cases and possible errors.
The supported file formats include both vector (e.g., [GeoPackage](https://en.wikipedia.org/wiki/GeoPackage), [Shapefile](https://en.wikipedia.org/wiki/Shapefile), [GeoJSON](https://geojson.org/)) and raster (e.g., [GeoTIFF](https://en.wikipedia.org/wiki/GeoTIFF), [JPEG 2000](https://en.wikipedia.org/wiki/JPEG_2000#GML_JP2_georeferencing)) file types.
The user can extract the spatial extent, the so called bounding box (parameter `bbox`), and/or temporal extent (parameter `tbox`) of a file or set of files.


### Individual files

geoextent can be run on a single data file using the Python API as shown below.
The output includes the spatial and temporal extents.
The spatial extent is always provided as in the [WGS84](https://en.wikipedia.org/wiki/World_Geodetic_System) coordinate reference system ([CRS](https://en.wikipedia.org/wiki/Spatial_reference_system)), commonly known though its usage in GPS, which is sufficiently precise for the use case of dataset discovery. If the dataset is provided in a different CRS, the bounding box is reprojected using GDAL.

```python
# File of interest
local_filepath = "../tests/testdata/shapefile/ifgi_denkpause.shp"
# Geoextent extraction
geoextent_file = geoextent.fromFile(filepath = local_filepath, bbox = True, tbox = True)

# Print output
print(local_filepath, "\n", "bbox:", geoextent_file['bbox'],"\n",geoextent_file['crs'], "EPSG \n",
     "tbox:",geoextent_file['tbox'])
```

### Multiple files

When provided a directory, geoextent by default returns the union of the spatial and temporal extent of all supported files.

```python
# Directory of interest
local_directory_path = "../tests/testdata/folders/folder_two_files"
# Geoextent extraction
geoextent_directory = geoextent.fromDirectory(path = local_directory_path, bbox = True, tbox = True, details = True)

# Print output
print(local_directory_path, "\n", "bbox:", geoextent_directory['bbox'],"\n", geoextent_directory['crs'], "EPSG \n",
     "tbox:",geoextent_directory['tbox'])
```

## Case study

Zenodo is a data repository that stores different types of publication materials as unique `records`. The Zenodo API allows the discovery of records by dozens of different parameters store in the metadata. These parameters includes of `title`, `doi`, `keywords`, and `locations`. Even though geospatial metadata is available as `locations` parameters, this options seems to be limited to only a small number of records. Therefore, we study the amount of records with geospatial metadata available in Zenodo and assess if by using geoextent we could provide the missing geospatial metadata for a significant number of some Zenodo records.

### Zenodo geometadata

The first step for this analysis is determining the current state of geometadata in the Zenodo records. As we are interested in analyzing some of these records with **geoextent**, only Open Access repositories are taken into account. For this purpose we are going to use the Zenodo __[spatial search](https://zenodo.org/api/records/?bounds=-180,-90,180,90)__ function available in their API.  This function requires the user to define an interest area (bounding box) to extract the records within this zone. Zenodo's geospatial information of a record is store in the `locations` property in the metadata. Each record could have multiple locations, but a location is limited to a single point. This point include a `lat`, `lon` and `place` features.

In the following cell we extract the proportion of open access Zenodo records with this `locations` property available. In this sense, is important to notice that Zenodo records include 9 different categories of publication types: `poster`, `presentation`, `dataset`, `image`, `video`, `software`, `lesson`, `physicalobject` and `other`. 

```python
## Parameters for search
world_bounds = "-180,-90,180,90"

# Type of Records
types_of_records = ["publication","poster", "presentation", "dataset", "image",
                    "video", "software", "lesson", "physicalobject","other"]

# Dictionary with search parameters
search_param = {}

# Zenodo search parameter for open records
search_param["open_record"] =  {"access_right":"open"}

# Zenodo search parameter for open records with geometadata
search_param["open_record_geometa"] =  dict(search_param["open_record"], bounds= world_bounds)

# Zenodo search parameter for open records with and without geometadata by type of record
for record_type in types_of_records:
    open_record = 'open_'+ record_type
    open_geometadata = open_record + "_geometa"
    search_param[open_record] = {"access_right":"open","type":record_type}
    search_param[open_geometadata] = {"access_right":"open","type":record_type,"bounds":world_bounds}
    
## Extract the number of records by search_parameter

num_of_records_by_search = {}

for type_of_zenodo_record in list(search_param.keys()):
    num_of_records_by_search[type_of_zenodo_record] = zenodo_api.get_number_of_records(search_param[type_of_zenodo_record])
    
```

Base on the information extracted from Zenodo's API, there are currently **{{'{:,}'.format(num_of_records_by_search['open_record'])}}** open records. From all open records, **{{'{:,}'.format(num_of_records_by_search["open_record_geometa"])}}** have geospatial metadata. This means that **{{'{:.2f}'.format(num_of_records_by_search["open_record_geometa"]/num_of_records_by_search['open_record']*100)}}%** open Zenodo records are findable throught geospatial search. Among the records with geospatial metadata, **{{'{:.2f}'.format(num_of_records_by_search["open_publication_geometa"]/num_of_records_by_search["open_record_geometa"]*100)}}%** are`publications`({{'{:,}'.format(num_of_records_by_search['open_publication_geometa'])}} records) and **{{'{:.2f}'.format(num_of_records_by_search["open_dataset_geometa"]/num_of_records_by_search["open_record_geometa"]*100)}}%** are of `dataset` type ({{'{:,}'.format(num_of_records_by_search["open_dataset_geometa"])}} records). Even for these 2 categories with most of the records, 
only **{{'{:.2f}'.format(num_of_records_by_search["open_publication_geometa"]/num_of_records_by_search["open_publication"]*100)}}%** of open `publications` and **{{'{:.2f}'.format(num_of_records_by_search["open_dataset_geometa"]/num_of_records_by_search["open_dataset"]*100)}}%** open `dataset` include geospatial metadata.

To evaluate if we can use **geoextent** to increase the proportion of records with spatial metadata we use the result of a Zenodo API term search to create a list of records to analyze. In this particular search, we use the term *{{SEARCH_TERM}}* and limit it to only `dataset` records smaller than 75 mb. From these records we extract basic metadata (e.g Title, DOI).

```python
# Parameters search for datasets with 'geo' term
geo_datasets_param = dict(search_param["open_dataset"], q = SEARCH_TERM)
df_geo_zenodo_api = pd.DataFrame.from_dict(zenodo_api.get_info_records(geo_datasets_param),orient= 'index')
# Selecting records smaller than 75 mb.
df_geo_zenodo_api = df_geo_zenodo_api[df_geo_zenodo_api["size_mb"]<MAXIMUM_RECORD_SIZE_MB].copy()
# Save API results
df_geo_zenodo_api.to_csv(OUTPUT_API)
```

## Analysis

The first one with the information of records (**repositories**) and the second for individual **files**.

```python
# Loads results from api and geoextent extraction
df_geo_zenodo_api = pd.read_csv(OUTPUT_API, index_col=0)
df_geo_zenodo_api.index = df_geo_zenodo_api.index.astype(str)
df_geo = pd.read_csv(OUTPUT_GEOEXTENT, index_col=0)

# Function to transform wkt into geometry
def wkt_loads(x):
    try:
        return wkt.loads(x)
    except:
        return None
    
# Transformats wkt into geometry for the geoextent extraction results
df_geo['bbox'] = df_geo.bbox.apply(wkt_loads)

# Creates two different geodataframes, one for files and other for zenodo records (repositories)
gdf_repository = gpd.GeoDataFrame(df_geo[df_geo['handler'] == "geoextent:"+geoextent_version],geometry ="bbox",crs = 4326)
gdf_repository.drop(['filename'], axis = 1,inplace = True)
gdf_files = gpd.GeoDataFrame(df_geo[df_geo['handler'] != "geoextent:"+geoextent_version],geometry ="bbox",crs = 4326)
```

### Geoextent extraction

To initially understand the performance of geoextent extracting geospatial and temporal information from zenodo records, we assess the general results after the extraction. There are three possible outcomes: 1) A successful extraction, 2) no information extracted, or 3) a failure ("error") of geoextent during the extraction. As it is possible to extract both the geospatial and temporal, just one of them or none, we analyze them individually.

```python
# Extract statistics for geospatial extraction 
num_records = len(gdf_repository)
per_records_geoextent = sum(gdf_repository.bbox.is_valid)/num_records*100
per_records_no_geoextent = (sum(gdf_repository.format == "repository")/num_records*100) - per_records_geoextent
per_records_failure = sum(gdf_repository.format == "repository_error")/num_records*100

# Extract statistics for temporal extraction
per_records_temp = (num_records-sum(gdf_repository.tbox.isnull()))/num_records*100
per_records_no_temp = (sum(gdf_repository.tbox.isnull())-sum(gdf_repository.format == "repository_error"))/num_records*100

# Store results
rec_successful = [per_records_geoextent,per_records_temp]
rec_no_extraction = [per_records_no_geoextent,per_records_no_temp]
rec_failure =  [per_records_failure,per_records_failure]

## Plot
fig, ax = plt.subplots(figsize=(15,3))

# Plot configuration
ext_type = ["Geospatial \n extraction","Temporal \n extraction"]
plt.title("Geoextent extraction of Zenodo records by parameter")
plt.barh(ext_type, rec_successful,color='g',label ='Successful')
plt.barh(ext_type, rec_no_extraction,color='grey',label ='No extraction',left = rec_successful)
plt.barh(ext_type, rec_failure,color='red',label ='Geoextent failure',left = [sum(x) for x in zip(*[rec_successful,rec_no_extraction])])
plt.xlabel('(%) Percentage of repositories')
fig_num = 1
plt.annotate('Figure '+ str(fig_num)+'. Proportion of extraction status for zenodo records',
             (0,0), (200, -40), xycoords='axes fraction', weight='bold', textcoords='offset points', va='center')
plt.annotate('Extracted with geoextent \nhttps://o2r.info/geoextent/', (0,0), (710,-20), xycoords='axes fraction', textcoords='offset points', va='top')
fig_num +=1
plt.legend(loc=5)
plt.show()
```

From a total of **{{num_records}}** Zenodo records analyzed we extracted **{{'{:.2f}'.format(per_records_geoextent)}}%** of their geospatial component and  **{{'{:.2f}'.format(per_records_temp)}}%** their temporal one. Geoextent did not retrieve the geospatial and temporal extent in **{{'{:.2f}'.format(per_records_no_geoextent)}}%** and **{{'{:.2f}'.format(per_records_no_temp)}}%** of the cases. That means that geoextent explored all files in those records, but it did not extract any geospatial or temporal component. Finally, the extraction with geoextent failed (crash) due to undetermined reasons in **{{'{:.2f}'.format(per_records_failure)}}%** of the records.

A similar analysis is possible for the individual files. In this case, only two outputs are possible: either the extraction is successful, or there is no extraction at all. First, as geoextent stores the combined result of folders and zip files, we need to remove this type of registers to avoid overestimating the percentage of successful extraction.

```python
# Extract only files (no folders or zipfile)
uni_files_gdf = gdf_files[~gdf_files.format.isin(["folder","zip"])].copy().reset_index(drop=True)
uni_files_gdf.format = uni_files_gdf.format.str.lower()

# Extract statistics for geospatial extraction 
num_files = len(uni_files_gdf)

# Percentage files with valid bbox (i.e., valid geometry) from total number of files
per_files_geoextent = sum(uni_files_gdf.bbox.is_valid)/num_files*100
# Percentage files without valid bbox (i.e., no geometry) from total number of files
per_files_no_geoextent = 100-per_files_geoextent

# Extract statistics for temporal extraction
per_files_temp = (num_files-sum(uni_files_gdf.tbox.isnull()))/num_files*100
per_files_no_temp = 100-per_files_temp

# Record results
files_successful = [per_files_geoextent,per_files_temp]
files_no_extraction = [per_files_no_geoextent,per_files_no_temp]

# Plot
status = ["Successful","No extraction"]
fig, ax = plt.subplots(figsize=(15,3))
plt.title("Geoextent extraction of files by parameter")
plt.barh(ext_type, files_successful,color='g',label ='Successful')
plt.barh(ext_type, files_no_extraction,color='grey',label ='No extraction',left = files_successful)
plt.xlabel('(%) Percentage of files')
plt.xlim(0,100)
plt.legend(status,loc=5)
plt.annotate('Figure '+ str(fig_num)+'. Proportion of success and no extraction by individual files', (0,0), (250, -40), xycoords='axes fraction', weight='bold', textcoords='offset points', va='top')
plt.annotate('Extracted with geoextent \nhttps://o2r.info/geoextent/', (0,0), (710,-20), xycoords='axes fraction', textcoords='offset points', va='top')
fig_num+=1
plt.show()
```

From **{{'{:,}'.format(num_files)}}** files analyzed, **{{'{:.2f}'.format(per_files_temp)}}%** had a successful temporal extraction and **{{'{:.2f}'.format(per_files_geoextent)}}%** a geospatial extraction. Files with no extraction of temporal or geospatial extraction are a combination of different types of files. For instance, the zenodo records could include information that does not necessarily includes (or require) temporal or geospatial features as laboratory results, code, or text-based documents. Similarly, even for file formats designed to store geospatial data, there is the risk of ambiguity or lack of information regarding the coordinate reference systems. Lastly, even though geoextent relies on GDAL that supports hundreds of raster and vector drivers, there could be file formats not supported by geoextent.

In this sense, we took those files with known formats to potentially store geospatial information and classified them as files with potential geoextent. This group gives us a better idea of the proportion of geospatial metadata that we could extract using geoextent for zenodo records.

```python
# List of potential file formats with geospatial information
potential_geo_formats = ['csv','nc','tif', 'dbf',
                               'shp', 'shx', 'gpkg', 'geojson', 
                               'png','gml','xml','pdf','jpeg','jpg','qpj',
                               'png','json']


# Extract total number of files by zenodo record 
total_files = uni_files_gdf.zenodo_record_id.value_counts()

# Extract total number of potential file with geoextent by zenodo record
files_potential_geo_formats = uni_files_gdf['format'].isin(potential_geo_formats).groupby(uni_files_gdf['zenodo_record_id']).sum()

# Extract total number of files with potential geospatial information
total_files_potential = sum(files_potential_geo_formats)

# Extract percentage of files with successful geoextent extraction from the potential
per_total_geoextent_potential = sum(uni_files_gdf.bbox.is_valid)/total_files_potential*100

## Plot
fig, ax = plt.subplots(figsize=(15,3))
# Plot configuration
plt.title("Geoextent extraction status by potential files with geoextent")
plt.barh("Geoextent \n extraction",per_total_geoextent_potential,color='g',label ='Successful extraction')
plt.barh("Geoextent \n extraction",100-per_total_geoextent_potential,color='grey',label ='No extraction',left = per_total_geoextent_potential)
plt.xticks(np.arange(0,101, 5.0))
plt.xlabel('(%) Percentage of files')
plt.annotate('Figure '+str(fig_num)+'. File status extraction for files \n with potential geospatial information', (0,0), (300, -40), xycoords='axes fraction', weight='bold', textcoords='offset points', va='top')
plt.annotate('Extracted with geoextent \nhttps://o2r.info/geoextent/', (0,0), (710,-20), xycoords='axes fraction', textcoords='offset points', va='top')
fig_num +=1
plt.legend()
plt.show()
```

From **{{'{:,}'.format(num_files)}}** files, there are **{{'{:,}'.format(total_files_potential)}}** files with known formats that could potentially store geospatial information. From these files, only **{{'{:.2f}'.format(per_total_geoextent_potential)}}%** had a successful geoextent extraction (i.e., BBOX extracted) and **{{'{:.2f}'.format(100-per_total_geoextent_potential)}}%** resulted in no extraction. The distribution of these files among the repositories is relevant to explore the percentages of zenodo records that could have a successful geoextent extraction (i.e., at least one file with successful extraction).

```python
# Extract total number of files with successful geoextent extraction by zenodo record
geo_files = uni_files_gdf['bbox'].is_valid.groupby(uni_files_gdf['zenodo_record_id']).sum()

#Extract number of records by category
num_records = len(files_potential_geo_formats)
per_potential = sum(files_potential_geo_formats>0)/num_records*100
per_potential_geoextent = sum(geo_files>0)/num_records*100
per_potential_no_geoextent = per_potential-per_potential_geoextent

## Plot
fig, ax = plt.subplots(figsize=(15,3))
# Plot configuration
plt.title("Potential geoextent by zenodo records")
plt.barh("Records with successful \n geoextent extraction",per_potential_geoextent,color='green',edgecolor="black",hatch = '-',alpha=0.7,label ='No extraction')
plt.barh("Records with potential \n but no geoextent",per_potential_no_geoextent,color= 'r',edgecolor="black",hatch = '|',alpha=0.5,label ='No extraction')
plt.barh("Records with \n potential geoextent",per_potential,alpha=0.5, color = 'orange',edgecolor="black",hatch = '+',label ='Successful extraction')
plt.annotate('Figure '+str(fig_num)+'. Percentage of zenodo records with potential geoextent,\n potential but no geoextent or successful geoextent extraction.', (0,0), (200, -40), xycoords='axes fraction', weight='bold', textcoords='offset points', va='top')
plt.annotate('Extracted with geoextent \nhttps://o2r.info/geoextent/', (0,0), (710,-20), xycoords='axes fraction', textcoords='offset points', va='top')
plt.xlabel('(%) Percentage of zenodo records')
fig_num +=1
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
plt.show()
```

From the **{{'{:,}'.format(num_records)}}** zenodo records analyzed (i.e., total records without taking into account **{{'{:,}'.format(sum(gdf_repository.format == "repository_error"))}}** failed geoextent extractions), **{{'{:.2f}'.format(per_potential)}}%** of them have at least one file from a supported file format. Consequently, in the best scenario, the percentage of successful geospatial extraction would be **{{'{:.2f}'.format(per_potential)}}%**. However, in reality, only **{{'{:.2f}'.format(per_potential_geoextent)}}%** had a successful extraction and, **{{'{:.2f}'.format(per_potential_no_geoextent)}}%** had the potential to be successful (i.e., at least one file with supported file format) but resulted in no extraction.

```python
# Extract number of total files by records with successful extraction by zenodo record
files_in_record = uni_files_gdf.zenodo_record_id.value_counts()
# Extract number of files with successful extraction with potential geospatial information by zenodo record
potential_files_with_geo_in_record = uni_files_gdf['format'].isin(potential_geo_formats).groupby(uni_files_gdf['zenodo_record_id']).sum()
# Extract number of files with successful geoextent extraction by zenodo record
files_with_geo_in_record = uni_files_gdf['bbox'].is_valid.groupby(uni_files_gdf['zenodo_record_id']).sum()

# Store information
result = pd.DataFrame({'files_in_record': files_in_record,
                       'files_with_potential_geoextent': potential_files_with_geo_in_record,
                       'files_with_extracted_geoextent': files_with_geo_in_record,
                      })

result['per_success_from_total'] = result.files_with_extracted_geoextent/result.files_in_record*100
result['per_success_from_potential'] = result.files_with_extracted_geoextent/result.files_with_potential_geoextent*100
result['potential_per_success'] = result['files_with_potential_geoextent']/result['files_in_record']*100

# This line includes repositories with 0 potential files. Replace NaN to 0 
# All repos without potential files go to (0,0)
result.per_success_from_potential.fillna(0, inplace=True)

## Plot

fig, ax = plt.subplots(figsize=(15,8))

#plt.fill_between(np.arange(0.0, 100, 0.01), np.arange(0.0, 100, 0.01),101,
#                 alpha=0.2, color = 'red',hatch = '+',edgecolor="white")

scatter = plt.scatter(result['potential_per_success'],result['per_success_from_potential'],
                     c=result['per_success_from_total'], s=result['files_in_record'], cmap="copper",alpha= 0.7)

range_per_success_total = list(np.arange(0.0,120,20))
# produce a legend with the unique colors from the scatter
#legend1 = ax.legend(*scatter.legend_elements(num=range_per_success_total), title="% of success \n from total files", bbox_to_anchor=(1.11,0.6))
plt.colorbar(label="Percentage of successful extraction \n over total files in repository")

range_num_files = list(result['files_in_record'].quantile([0.1,0.3,0.6,0.8,0.9,0.99]).round(0))

# produce a legend with a cross section of sizes from the scatter
kw = dict(prop="sizes", num=range_num_files, fmt="{x:,}", func=lambda s: s,alpha=0.5)
legend2 = ax.legend(*scatter.legend_elements(**kw),title="# of total files \n in record" ,bbox_to_anchor=(0.15,0.99))
plt.xlabel('Percentage of files in repository with potential geospatial information')
plt.ylabel('Percentage of files with successful geospatial \n information extraction over potential files')
plt.annotate('Figure '+str(fig_num)+'. Distribution of files by Zenodo record', (0,0), (225, -40), xycoords='axes fraction', weight='bold', textcoords='offset points', va='top')
plt.annotate('Extracted with geoextent \nhttps://o2r.info/geoextent/', (0,0), (600,-20), xycoords='axes fraction', textcoords='offset points', va='top')
plt.annotate("1'",xy=(42, 20), xytext=(45, 25),arrowprops=dict(arrowstyle="->"))
plt.annotate("2'", xy=(99, 99), xytext=(92, 95),arrowprops=dict(arrowstyle="->"))

fig_num +=1
plt.show()
```

To understand which formats are problematic to extract geospatial information, we compute its percentage of extraction success.

```python
uni_files_gdf['format'].value_counts()
files_valid = uni_files_gdf[uni_files_gdf.geometry.is_valid].copy().reset_index(drop=True)

d1 = pd.DataFrame({"Total":uni_files_gdf['format'].value_counts(),
                  "geoextent_success":files_valid['format'].value_counts()})
d1.dropna(inplace=True)
d1['perc_success'] = (d1['geoextent_success']/d1['Total'])*100

labels = d1.index.values
success = d1['perc_success']
failure = 100 - d1['perc_success']

color_success = "#009900"
color_failure = "#FF3333"
fig, ax = plt.subplots(figsize=(15,4))
width = 0.35
plt.bar(labels, success, width, label='Successful',color= color_success)
plt.bar(labels, failure, width, bottom=success, color = color_failure,label='No extraction')
plt.ylabel('Percentage')
plt.xlabel('File format')
plt.title('Geoextent extraction success by supported file type',size = 20)
plt.ylabel('Percentage')

count = 0
for bar in ax.patches: 
    displace =-1
    color = color_failure
    pos = 90
    if bar.get_y() == 0:
        n = d1['Total'][count]
        count+=1
        displace = 3
        color= color_success
        pos = 0
        plt.annotate("n= "+str(n),  
                   (bar.get_x() + (bar.get_width()/2),  
                   50), ha='center', va='center', 
                   size=10, xytext=(0, 0),color='black', 
                   textcoords='offset points')
    plt.annotate(format(bar.get_height()/100, ".1%"),  
                   (bar.get_x() + (bar.get_width()*1.5),  
                   pos), ha='center', va='center', 
                   size=10, xytext=(0, 8),color=color, 
                   textcoords='offset points')
    
plt.legend(loc='center right', bbox_to_anchor=(1.1, 0.5))
plt.annotate('Figure '+str(fig_num)+'. Distribution of files by Zenodo record', (0,0), (280, -40), xycoords='axes fraction', weight='bold', textcoords='offset points', va='top')
plt.annotate('Extracted with geoextent \nhttps://o2r.info/geoextent/', (0,0), (710,-20), xycoords='axes fraction', textcoords='offset points', va='top')
fig_num +=1
plt.show()
```

First, we can observe that file formats whose main purpose is to store geospatial information have a high success rate of extraction. For example, GeoJSON files (JSON format for geospatial information) have a **{{'{:.2f}'.format(d1.loc['geojson'].perc_success)}}%** success rate and .tif, .shp and .dbf vary between **{{'{:.2f}'.format(d1.loc['dbf'].perc_success)}}%** and **{{'{:.2f}'.format(d1.loc['tif'].perc_success)}}%** of success rate. However, this is not always the case. For example, geo package (gpkg), an open format for geospatial information, has a lower success rate of **{{'{:.2f}'.format(d1.loc['gpkg'].perc_success)}}%**. That could result from the small sample size (n=**{{'{:,}'.format(d1.loc['gpkg'].Total)}}**) and errors in the formatting of files as those present in other geospatial file formats (e.g., **{{'{:.2f}'.format(100-d1.loc['shp'].perc_success)}}%** failure in the files).

Second, the results suggest that ambiguous formats (i.e., those that do not necessarily have geospatial information) have a high percentage of no extractions. For CSV files, geoextent could not extract geospatial information in **{{'{:.2f}'.format(100-d1.loc['csv'].perc_success)}}%** of files. To determine what percentage corresponds to information without geospatial data (e.g., laboratory result) and those with geospatial information but unsuccessful extraction, we would need to evaluate each file one by one. However, what is relevant for us is that at least **{{'{:.2f}'.format(d1.loc['csv'].perc_success)}}%** of CSV files have extractable geospatial information. Files with a .png format also have a similar low success rate of **{{'{:.2f}'.format(d1.loc['png'].perc_success)}}%**. That means that we can not discard ambiguous file formats as potential sources of geometada.
